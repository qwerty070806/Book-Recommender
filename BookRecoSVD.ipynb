```python
import numpy as np

# Let's use the simple 3x2 matrix from our previous example
# A = | 7  2 |
#     | 6  1 |
#     | 2  8 |
A = np.array([
    [1,1],
    [7,7]
])

# Perform Singular Value Decomposition
# The function returns U, a 1D array of singular values 's', and V transpose 'Vt'
U, s, Vt = np.linalg.svd(A)

# --- Print the results ---

print("Original Matrix (A):")
print(A)
print("-" * 30)

print("U (User-to-Feature Matrix):")
print(U)
print("-" * 30)

# s is a 1D array of singular values. We need to create the full Sigma matrix.
# The full Sigma matrix has the same shape as A, with the singular values on its diagonal.
Sigma = np.zeros(A.shape)
Sigma[:min(A.shape), :min(A.shape)] = np.diag(s)

print("Σ (Feature Strength Matrix):")
print(Sigma)
print("-" * 30)

print("Vᵀ (Book-to-Feature Matrix):")
print(Vt)
print("-" * 30)


# --- Verification: Reconstruct the original matrix ---
# A_reconstructed = U @ Sigma @ Vt
A_reconstructed = np.dot(U, np.dot(Sigma, Vt))

print("Reconstructed Matrix (U @ Σ @ Vᵀ):")
print(A_reconstructed)
print("\nNote: The reconstructed matrix is identical to the original matrix.")
```

```python
from surprise import Dataset, Reader, SVD
from surprise.model_selection import train_test_split
from surprise import accuracy
```

```python
# Load the dataset using the Reader
reader = Reader(rating_scale=(0, 10))
data = Dataset.load_from_df(final_ratings[['User-ID', 'Book-Title', 'Book-Rating']], reader)
final_ratings_explicit = final_ratings[final_ratings['Book-Rating'] > 0]
data_explicit = Dataset.load_from_df(final_ratings_explicit[['User-ID', 'Book-Title', 'Book-Rating']], reader)

# Split the data into training and testing sets
trainset, testset = train_test_split(data_explicit, test_size=0.2, random_state=42)

# Train the SVD model
svd_model = SVD()
svd_model.fit(trainset)

# Make predictions on the test set
predictions = svd_model.test(testset)

# Add predictions to the original test DataFrame for evaluation
# Need to convert surprise predictions back to a pandas DataFrame
predicted_ratings_df = pd.DataFrame(predictions, columns=['User-ID', 'Book-Title', 'actual_rating', 'predicted_rating', 'details'])

# Merge predicted ratings back to the original final_ratings test set
# This requires re-creating the test set from the original final_ratings based on the split
# A simpler approach is to use the testset from surprise which contains the original ratings

# We can calculate RMSE directly from the predictions object
rmse = accuracy.rmse(predictions)

print("Root Mean Squared Error (RMSE) for SVD:", rmse)


# To display predictions in a DataFrame similar to previous steps,
# we can convert the predictions list to a DataFrame and merge if needed,
# but for this subtask, just showing the RMSE is sufficient as per instructions.
# If a display is required, we can do:
display(predicted_ratings_df)
```

```python
from surprise.model_selection import cross_validate

# Use the same 'data_explicit' from the step above
svd = SVD()
cv_results = cross_validate(svd, data_explicit, measures=['RMSE', 'MAE'], cv=5, verbose=True)

print("\nMean RMSE from Cross-Validation:", cv_results['test_rmse'].mean())
```

```python
from surprise import SVD
from surprise.model_selection import GridSearchCV, train_test_split
from surprise import accuracy

# Parameter grid for SVD
param_grid = {
    'n_factors': [50, 100, 150],   # number of latent factors
    'n_epochs': [20, 30, 50],      # number of training iterations
    'lr_all': [0.002, 0.005, 0.007], # learning rate
    'reg_all': [0.02, 0.05, 0.1]   # regularization term
}

# Grid search
gs = GridSearchCV(SVD, param_grid, measures=['rmse'], cv=3, n_jobs=-1, joblib_verbose=2)
gs.fit(data_explicit)

# Best RMSE and parameters
print('Best RMSE from CV:', gs.best_score['rmse'])
print('Best parameters:', gs.best_params['rmse'])

# Train model with best parameters
best_svd = gs.best_estimator['rmse']
trainset, testset = train_test_split(data_explicit, test_size=0.2, random_state=42)
best_svd.fit(trainset)
predictions = best_svd.test(testset)

# Evaluate final tuned model
final_rmse = accuracy.rmse(predictions)
print("Final tuned RMSE:", final_rmse)
```

```python
def get_top_n_recommendations(user_id, model, data, n=10):
    """
    Generates top N book recommendations for a given user using a trained SVD model.

    Args:
        user_id (int): The ID of the user to get recommendations for.
        model (surprise.prediction_algorithms.matrix_factorization.SVD): The trained SVD model.
        data (pd.DataFrame): The dataframe containing 'User-ID', 'Book-Title', and ratings.
        n (int): The number of recommendations to return.

    Returns:
        list: A list of tuples, where each tuple contains (book_title, predicted_rating).
              Returns an empty list if the user_id is not found.
    """
    # 1. Check if the user exists in the dataset.
    if user_id not in data['User-ID'].unique():
        print(f"Error: User ID {user_id} not found in the dataset.")
        return []

    # 2. Get the list of all unique book titles.
    all_book_titles = data['Book-Title'].unique()

    # 3. Get the list of book titles the user has already rated.
    rated_books = data[data['User-ID'] == user_id]['Book-Title'].tolist()

    # 4. Create a list of books the user has NOT rated.
    unrated_books = [book for book in all_book_titles if book not in rated_books]

    # 5. Predict the user's rating for each unrated book.
    predictions = [model.predict(user_id, book) for book in unrated_books]

    # 6. Sort the predictions by the estimated rating in descending order.
    predictions.sort(key=lambda x: x.est, reverse=True)

    # 7. Get the top N predictions.
    top_n_predictions = predictions[:n]

    # 8. Format the output as a list of (book_title, predicted_rating) tuples.
    top_n_recommendations = [(pred.iid, pred.est) for pred in top_n_predictions]

    return top_n_recommendations

# --- USAGE EXAMPLE ---
# You can now use this function with your best-trained model.
# (Assuming 'best_svd' and 'final_ratings_explicit' are already defined from previous steps)

target_user_id = 11676
recommendations = get_top_n_recommendations(target_user_id, best_svd, final_ratings_explicit, n=10)

if recommendations:
    print(f"\nTop 10 book recommendations for User ID: {target_user_id}")
    for book, rating in recommendations:
        print(f"- {book} (Predicted Rating: {rating:.2f})")
```

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
padhe_likhe_users = x[x].index #experienced users
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe_users)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
final_ratings_explicit = final_ratings[final_ratings['Book-Rating'] > 0]
```

```python
# Create a pivot table: Book-Title as index, User-ID as columns, and Book-Rating as values
pt = final_ratings_explicit.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')

# Fill missing values with 0. This is important for the similarity calculation.
pt.fillna(0, inplace=True)

# Display the pivot table
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
def recommend(book_name):
    # Check if the book exists in the pivot table index
    if book_name not in pt.index:
        print(f"Book '{book_name}' not found in the dataset for recommendations.")
        return []

    # index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:6]

    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))

        data.append(item)

    return data
```
